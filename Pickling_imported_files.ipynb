```python
import pandas as pd

import pickle

import glob

import re
```

The data is available at https://www.nhtsa.gov/research-data/fatality-analysis-reporting-system-fars

The csvs are held within a folder called fatalitydata/ located in the parent directory of the repo so as not to clutter it.

```python
df_name_list = []

path =r'../fatalitydata'
allFiles = glob.glob(path + "/*.csv")
regex = re.compile('(?<=/)[\w]+(?=\.csv)')
for file_path in allFiles:
    print(file_path)
    file_name = re.findall(regex,file_path)[0]
    df_name_list.append("df_"+file_name)
    exec("df_"+file_name+" = pd.read_csv(file_path,index_col=None, header=0,encoding = 'ISO-8859-1',error_bad_lines=False)")
```

```python
def df_pickler(df_name_list):
    for _df in df_name_list:
        exec(_df+".to_pickle(path='./pickles_for_project/" + _df + ".pkl')")
    with open('./pickles_for_project/list_df_names.pkl', 'wb') as f:
        pickle.dump(df_name_list, f)
```
